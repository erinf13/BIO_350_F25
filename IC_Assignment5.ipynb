SIR = Susceptible, Infected, Removed (dead in this example)

St+1 = St - (beta * St * It)

It+1 = It + (beta * St * It) - (d * It)

Rt+1 = Rt + (d * It)

St+1 + It+1 + Rt+1 = St + It + Rt


    F0 = 1000 (initial population size)
    S0 = 999 (initial susceptible)
    I0 = 1 (initial infected)
    R0 = 0 (initial removed)
    d = 0.5 (probability of death per day)
    beta = 0.001 (probability of transmission per contact per day)

run the model for 30 days and plot the results using for loops

set return variables 
set initial variables 
insert equations into for loop
giving the function beta, d, S0, I0, R0, and F0, and the number of days to run the model (t = 30)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create a function that takes initial S, I, R values, death rate, beta, and time as inputs
def SIR_model(S0, I0, R0, beta, d, t):
    """
    Parameters:
    S0 = 999 (initial susceptible)
    I0 = 1 (initial infected)
    R0 = 0 (initial removed)
    d = 0.5 (probability of death per day)
    beta = 0.001 (probability of transmission per contact per day)
    t = 30 (number of days to simulate)
    """
    #import the variables given to the function
    #set up lists to hold S, I, R values over time
    Svalues = [S0]
    Ivalues = [I0]
    Rvalues = [R0]

    #set current values to initial values
    currentS = S0
    currentI = I0
    currentR = R0

    #for loop to go through time steps
    for i in range(t):
        #calculate next S, I, R values
        currentS = currentS - (beta * currentS * currentI)
        currentI = currentI + (beta * currentS * currentI) - (d * currentI)
        currentR = currentR + (d * currentI)

        #append next values to lists and round to 3 decimal places
        Svalues.append(round(currentS, 3))
        Ivalues.append(round(currentI, 3))
        Rvalues.append(round(currentR, 3))
    return Svalues, Ivalues, Rvalues

```

```python
#Call the SIR_model function with the parameters given in the docstring and save the output to variables
S, I, R = SIR_model(999, 1, 0, 0.001, 0.5, 30)
print(S)
print(I)
print(R)
```

```python
#plot the results
plt.plot(S, label='Susceptible')
plt.plot(I, label='Infected')
plt.plot(R, label='Removed')
plt.xlabel('Days')
plt.ylabel('Number of Individuals')
plt.title('SIR Model of Rabies in Fox Population')
plt.legend()
plt.show()
```

Questions:
1. What happenes when beta is changed?
- When increasing beta, the rate of infection increases, leading to a higher number of infected individuals and a faster spread of the disease, seen in the steeper curve of susceptible individuals decreasing. The infected population peaks higher and earlier, while the removed population also increases more rapidly as more individuals succumb to the disease.
- In contrast, decreasing beta would result in a slower spread of disease with fewer individuals infected at any given time, leading to a more gradual decline in the susceptible population and a lower peak in the infected population.
2. What happens when d is changed?
- Decreasing the death rate resulted in a higher number of infected individuals over time, as more foxes survive the infection longer, leading to a slower increase in the removed population. The susceptible population declines more gradually, and the peak of the infected population is lower and occurs later.
- Conversely, increasing the death rate causes a rapid increase in the removed population as more infected foxes die quickly, leading to a sharper decline in the susceptible population and a higher peak in the infected population that occurs earlier.
3. Why do you think these parameters have the effects they do?
- The parameters beta and death rate directly influence the dynamics of disease transmission and death within the population. Beta represents the transmission rate of the disease, so when it is higher, the infected fox is more likely to come into contact with susceptible foxes to spread the disease, which leads to the disease spreading more quickly (and vice versa). The death rate of infected individuals represents how quickly infected foxes die from rabies, so a higher death rate means the infected foxes die quickly, reducing the number of infectious individuals and slowing the spread of the disease (and vice versa). Therefore, changes in these parameters significantly affect how quickly the disease spreads and how many individuals ultimately succumb to it.

- Yes, the disease could eventually spread through the population. In the initial scenario with beta = 0.001 and d = 0.5, the model predicts that a significant portion of the fox population would eventually become infected and die from rabies over the 30-day period.
