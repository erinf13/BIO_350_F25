```python
import matplotlib.pyplot as plt
# install packages first
```

```python
def N_exponential(N_0, p, time):
    # Calculate population size over time using exponential growth model
    # save N values in a list
    N_values = [N_0]
    # save initial population
    current_N = N_0
    for _ in range(time):
        current_N = current_N + p * current_N
        # round to 3 decimal places
        current_N = round(current_N, 3)
        N_values.append(current_N)
    return N_values 
```

```python
#Run the function with 35 days and p = 0.4 with an initial population of 0.01
result = N_exponential(0.01, 0.4, 35)
print(result)  # This will print the population sizes over 35 days
```

```python
#Plot the results with days on the x-axis and population on the y-axis
days = list(range(36))  # Days from 0 to 35
# Create the plot
plt.plot(days, result, marker='o')
# Add labels and title
plt.xlabel('Days') 
plt.ylabel('Duckweed Population')
plt.title('Duckweed Population Growth Over Time')  
plt.grid()
plt.show()
```

```python
#plot the results again but log the y-axis variable
days = list(range(36))  # Days from 0 to 35
# Create the plot
plt.plot(days, result, marker='o')
plt.yscale('log')
plt.xlabel('Days') 
plt.ylabel('Log Duckweed Population')
plt.title('Duckweed Population Growth Over Time')  
plt.grid()
plt.show()
```

```python
#Change N_Exp to N_exp2 and have 3 conditions based on N_0
#If N_0 < 500, p = 0.4
#If N_0 >= 500 and N_0 < 1000, p = 0.2
#If N_0 >= 1000, p = 0

def N_exponential2(N_0, time):
    # exponential growth model with variable p based on N_0
    # save N values in a list
    N_values = [N_0]
    # save initial population
    current_N = N_0
    for _ in range(time):
        if current_N < 500:
            p = 0.4
        elif 500 <= current_N < 1000:
            p = 0.2
        else:
            p = 0
        current_N = current_N + p * current_N
        # round to 3 decimal places
        current_N = round(current_N, 3)
        N_values.append(current_N)
    return N_values
```
