Paper: Impacts of sequential and simultaneous coinfection by major amphibian pathogens on disease outcomes 

Faust E, T Carvalho, F De Jesus Andino, J Robert, TY James. 2025. Impacts of sequential and simultaneous coinfection by major amphibian pathogens on disease outcomes. Ecosphere 16(8): e70365. https://doi.org/10.1002/ecs2.70365. 

Data available at : https://datadryad.org/dataset/doi:10.5061/dryad.70rxwdc71#readme. 

They have all of their variables laid out in their readme and all code documented in an R script. 
They assessed the differences in host mortality among treatments by building Kaplan-Meier survival curves for each treatment group using the survival package in R, then compared the results using log-rank tests. Post hoc tests were performed using Tukey's post hoc tests and a series of linear regression models were created to assess differences in body condition. Paired t-tests were used to determine differences in pathogen load and body condition between time points.
