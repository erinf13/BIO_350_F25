```python
# Survival
import pandas as pd
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt

# Load your dataset
ce_perfrog = pd.read_csv("ce_perfrog.csv")

# Initialize Kaplan-Meier fitter
kmf = KaplanMeierFitter()

# --- Survival by unmodified death day ---
plt.figure(figsize=(8,6))

for treatment, group in ce_perfrog.groupby("treatment"):
    # Remove missing values in relevant columns
    df = group.dropna(subset=["death_day", "censor"])
    
    if len(df) > 0:
        kmf.fit(
            durations=df["death_day"], 
            event_observed=df["censor"], 
            label=treatment
        )
        kmf.plot_survival_function()

plt.title("Survival by Treatment (Death Day)")
plt.xlabel("Days")
plt.ylabel("Survival Probability")
plt.legend(title="Treatment")
plt.show()

# --- Survival POST-Bd ---
plt.figure(figsize=(8,6))

for treatment, group in ce_perfrog.groupby("treatment"):
    # Remove missing values for post-Bd data
    df = group.dropna(subset=["death_post_Bd", "censor"])
    
    if len(df) > 0:
        kmf.fit(
            durations=df["death_post_Bd"], 
            event_observed=df["censor"], 
            label=treatment
        )
        kmf.plot_survival_function()

plt.title("Survival by Treatment (Post-Bd)")
plt.xlabel("Days Post-Bd")
plt.ylabel("Survival Probability")
plt.legend(title="Treatment")
plt.show()
```

```python
# survival post hoc stats
import pandas as pd
from lifelines.statistics import logrank_test, multivariate_logrank_test, pairwise_logrank_test

# Load data
ce_perfrog = pd.read_csv("ce_perfrog.csv")

# Drop rows with missing values in relevant columns
ce_perfrog = ce_perfrog.dropna(subset=["death_day", "censor", "treatment"])

# --- GLOBAL survival comparison across all treatments (death_day) ---
survival = multivariate_logrank_test(
    event_durations=ce_perfrog["death_day"],
    groups=ce_perfrog["treatment"],
    event_observed=ce_perfrog["censor"]
)
print("Global survival comparison (death_day):")
print(survival.summary)
print("\n")

# --- PAIRWISE survival comparison (death_day) ---
pairwise_results = pairwise_logrank_test(
    event_durations=ce_perfrog["death_day"],
    groups=ce_perfrog["treatment"],
    event_observed=ce_perfrog["censor"]
)
print("Pairwise p-values (death_day):")
print(pairwise_results.p_value)

# Save to dataframe if you want
import itertools

# Extract treatment group names
groups = ce_perfrog["treatment"].unique()

# Build all pairwise combinations
pairs = list(itertools.combinations(groups, 2))

# Make a DataFrame of results
table_survival = pd.DataFrame({
    "Group 1": [p[0] for p in pairs],
    "Group 2": [p[1] for p in pairs],
    "p-value": pairwise_results.p_value.flatten()
})


# --- POST-Bd subset (exclude Rv treatment) ---
post_Bd = ce_perfrog[ce_perfrog["treatment"] != "Rv"].dropna(subset=["death_post_Bd"])

# Global log-rank test for post-Bd survival
surv_pBd = multivariate_logrank_test(
    event_durations=post_Bd["death_post_Bd"],
    groups=post_Bd["treatment"],
    event_observed=post_Bd["censor"]
)
print("\nGlobal survival comparison (Post-Bd):")
print(surv_pBd.summary)

# Pairwise log-rank comparisons for post-Bd survival
post_Bd_pairwise = pairwise_logrank_test(
    event_durations=post_Bd["death_post_Bd"],
    groups=post_Bd["treatment"],
    event_observed=post_Bd["censor"]
)
print("\nPairwise p-values (Post-Bd):")
print(post_Bd_pairwise.p_value)

# Save to dataframe if desired
groups_post = post_Bd["treatment"].unique()
pairs_post = list(itertools.combinations(groups_post, 2))

table_post_Bd_surv = pd.DataFrame({
    "Group 1": [p[0] for p in pairs_post],
    "Group 2": [p[1] for p in pairs_post],
    "p-value": post_Bd_pairwise.p_value.flatten()
})
```

```python
# endpoint body condition normality checks and correction
import pandas as pd
import numpy as np
from scipy import stats

# Load your dataset
ce_perfrog = pd.read_csv("ce_perfrog.csv")

# --- Extract endpoint fitness data by treatment ---
end_SMI_ctr  = ce_perfrog.loc[(ce_perfrog["treatment"] == "ctr")  & (ce_perfrog["fitness_end"].notna()), "fitness_end"]
end_SMI_Rv   = ce_perfrog.loc[(ce_perfrog["treatment"] == "Rv")   & (ce_perfrog["fitness_end"].notna()), "fitness_end"]
end_SMI_Bd   = ce_perfrog.loc[(ce_perfrog["treatment"] == "Bd")   & (ce_perfrog["fitness_end"].notna()), "fitness_end"]
end_SMI_RvBd = ce_perfrog.loc[(ce_perfrog["treatment"] == "RvBd") & (ce_perfrog["fitness_end"].notna()), "fitness_end"]
end_SMI_BdRv = ce_perfrog.loc[(ce_perfrog["treatment"] == "BdRv") & (ce_perfrog["fitness_end"].notna()), "fitness_end"]
end_SMI_sim  = ce_perfrog.loc[(ce_perfrog["treatment"] == "Bd+Rv") & (ce_perfrog["fitness_end"].notna()), "fitness_end"]

# --- Shapiro-Wilk normality tests ---
print("Normality checks (ShapiroWilk):")
for name, data in {
    "ctr": end_SMI_ctr,
    "Bd": end_SMI_Bd,
    "BdRv": end_SMI_BdRv,
    "Bd+Rv": end_SMI_sim,
    "RvBd": end_SMI_RvBd,
    "Rv": end_SMI_Rv
}.items():
    stat, p = stats.shapiro(data)
    print(f"{name}: W={stat:.3f}, p={p:.4f}")

# --- Log transform failed groups and recheck ---
print("\nAfter log transform:")
for name, data in {
    "Bd+Rv": end_SMI_sim,
    "RvBd": end_SMI_RvBd
}.items():
    stat, p = stats.shapiro(np.log(data))
    print(f"{name} (log): W={stat:.3f}, p={p:.4f}")

# --- Remove outliers (like in R) ---
# Sort and remove extremes
end_SMI_sim = np.sort(end_SMI_sim)[1:]    # remove smallest value
end_SMI_RvBd = np.sort(end_SMI_RvBd)[:6]  # keep first 6

# Recheck normality
print("\nAfter removing outliers:")
for name, data in {
    "Bd+Rv": end_SMI_sim,
    "RvBd": end_SMI_RvBd
}.items():
    stat, p = stats.shapiro(data)
    print(f"{name} (trimmed): W={stat:.3f}, p={p:.4f}")

# --- Remove outliers from main dataset for ANOVA ---
ce_perfrog_FIT = ce_perfrog.copy()

# Find and remove min from Bd+Rv
min_val = ce_perfrog.loc[(ce_perfrog["treatment"] == "Bd+Rv") & (ce_perfrog["fitness_end"].notna()), "fitness_end"].min()
min_idx = ce_perfrog.index[ce_perfrog["fitness_end"] == min_val][0]
ce_perfrog_FIT.loc[min_idx, "fitness_end"] = np.nan

# Find and remove max from RvBd
max_val = ce_perfrog.loc[(ce_perfrog["treatment"] == "RvBd") & (ce_perfrog["fitness_end"].notna()), "fitness_end"].max()
max_idx = ce_perfrog.index[ce_perfrog["fitness_end"] == max_val][0]
ce_perfrog_FIT.loc[max_idx, "fitness_end"] = np.nan

print(f"\nRemoved outliers at rows {min_idx} (Bd+Rv) and {max_idx} (RvBd)")
```

```python
# load difference between timepoints - T tests

import numpy as np
import pandas as pd
from scipy.stats import ttest_rel

def paired_ttest(df, treatment, var1, var2, label):
    """
    Perform a paired t-test on log-transformed values for a given treatment group.
    Prints summary results similar to R output.
    """
    x = np.log(df.loc[df["treatment"] == treatment, var1].dropna())
    y = np.log(df.loc[df["treatment"] == treatment, var2].dropna())

    # Ensure paired samples (same number of individuals)
    n = min(len(x), len(y))
    if n == 0:
        print(f"{label}: No data for treatment '{treatment}'")
        return None

    t_stat, p_val = ttest_rel(x[:n], y[:n])
    mean_diff = np.mean(x[:n] - y[:n])
    ci_low = mean_diff - 1.96 * (np.std(x[:n] - y[:n], ddof=1) / np.sqrt(n))
    ci_high = mean_diff + 1.96 * (np.std(x[:n] - y[:n], ddof=1) / np.sqrt(n))

    print(f"=== {label} ({treatment}) ===")
    print(f"n = {n}")
    print(f"t = {t_stat:.4f}, p = {p_val:.4g}")
    print(f"95% CI for difference: ({ci_low:.3f}, {ci_high:.3f})")
    print(f"Mean difference: {mean_diff:.3f}")
    if p_val < 0.05:
        print(" Significant difference (p < 0.05)")
    print("")

# --- Bd load comparisons ---
# 
paired_ttest(ce_perfrog, "Bd", "Bd_d10", "Bd_end", "Bd load difference")
paired_ttest(ce_perfrog, "BdRv", "Bd_d10", "Bd_end", "Bd  Rv load difference")
paired_ttest(ce_perfrog, "Bd+Rv", "Bd_d10", "Bd_end", "Bd+Rv load difference")
paired_ttest(ce_perfrog, "RvBd", "Bd_d20", "Bd_end", "Rv  Bd load difference")

# --- Rv load comparisons ---
paired_ttest(ce_perfrog, "Rv", "Rv_d10", "Rv_end", "Rv load difference")
paired_ttest(ce_perfrog, "RvBd", "Rv_d10", "Rv_end", "Rv->Bd load difference")
paired_ttest(ce_perfrog, "Bd+Rv", "Rv_d10", "Rv_end", "Bd+Rv load difference")
paired_ttest(ce_perfrog, "BdRv", "Rv_d10", "Rv_end", "Bd->Rv load difference")
```

```python
# survival since first exposure / Deathday model
import pandas as pd
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter

# ---- Load data ----
# make ce_perfrog a pandas DataFrame
# with columns:
#   death_day
#   censor (1 = died, 0 = censored)
#   treatment (factor)

ce_perfrog = pd.read_csv("ce_perfrog.csv")


# ---- Define colors (same as your R script) ----
colors = {
    "ctr": black,
    "Bd": blue,
    "Rv": orange,
    "BdRv": purple,
    "RvBd": green,
    "Bd+Rv": maroon
}

# Line types: R uses 1=solid, 2=dashed
linestyles = {
    "ctr": "dashed",
    "Bd": "solid",
    "Rv": "solid",
    "BdRv": "solid",
    "RvBd": "solid",
    "Bd+Rv": "solid"
}

linewidths = {
    "ctr": 1,
    "Bd": 2,
    "Rv": 2,
    "BdRv": 2,
    "RvBd": 2,
    "Bd+Rv": 2
}

# Order of groups in the plot (matches R)
group_order = ["ctr", "Bd", "Rv", "BdRv", "RvBd", "Bd+Rv"]

# ---- Fit KaplanMeier curves ----
km = KaplanMeierFitter()

plt.figure(figsize=(6,6))

for group in group_order:
    subset = ce_perfrog[ce_perfrog["treatment"] == group]

    km.fit(
        durations=subset["death_day"],
        event_observed=subset["censor"],
        label=group
    )
    
    km.plot(
        ci_show=False,
        color=colors[group],
        linestyle=linestyles[group],
        linewidth=linewidths[group]
    )

# ---- Formatting ----
plt.xlabel("Time (days after 1st exposure)", fontsize=12)
plt.ylabel("Survival Probability", fontsize=12)
plt.title("Survival Probability by Group", fontsize=16)
plt.xlim(0, 34)

# --- Custom Legend ---
plt.legend(
    [
        "ctr",
        "Bd",
        "Rv",
        "BdRv*",
        "RvBd*",
        "Bd+Rv",
    ],
    loc="upper center",
    bbox_to_anchor=(0.5, -0.1),
    ncol=3,
    fontsize=10,
    frameon=False
)

# ---- Add text annotations  ----
plt.text(7, 0.02, "*1st exposure (day 0)  2nd exposure (day 10)", fontsize=8)

plt.text(14, 0.15, "a", fontsize=12)
plt.text(18, 0.15, "a", fontsize=12)

plt.text(22, 0.10, "b", fontsize=12)
plt.text(26, 0.18, "b", fontsize=12)

plt.text(30, 0.95, "c", fontsize=12)
plt.text(31, 0.80, "c", fontsize=12)

plt.text(29, 0.50, "p < 0.002", fontsize=12)

plt.tight_layout()
plt.show()
```
