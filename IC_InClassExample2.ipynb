Importing and plotting population data.

```python
#Import "Population_Data.csv" into a pandas DataFrame
import pandas as pd
```

```python
#Import "Population_Data.csv" into a pandas DataFrame
pop_data = pd.read_csv("Population_Data.csv")
pop_data.head()
```

```python
# average the population from day 17 to day 60
pop_data[(pop_data['Days'] >= 17) & (pop_data['Days'] <= 60)]['Population'].mean()
print("The average population from day 17 to day 60 is", pop_data[(pop_data['Days'] >= 17) & (pop_data['Days'] <= 60)]['Population'].mean())
```

```python
#plot the population data with days on the x-axis and population size on the y-axis
import matplotlib.pyplot as plt
# insert grid, labels, and title
plt.grid(True)
plt.xlabel('Days')
plt.ylabel('Population Size')
plt.title('Population Growth Over Time')
# plot the data with points connected by lines and black color
plt.plot(pop_data['Days'], pop_data['Population'], marker='o', color='black', linestyle='-')
# add a horizontal line at y=276 to represent carrying capacity
plt.axhline(y=276, color='green', linestyle='--', label='y=276')
plt.legend()
# set x and y axis limits to start at 0
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.show()
```
